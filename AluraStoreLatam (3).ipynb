### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tienda["Precio"] = tienda["Precio"].round(0).astype(int)
tienda2["Precio"] = tienda2["Precio"].round(0).astype(int)
tienda3["Precio"] = tienda3["Precio"].round(0).astype(int)
tienda4["Precio"] = tienda4["Precio"].round(0).astype(int)

totalventa = tienda["Precio"].sum()
print(f"la venta en la tienda #1 es: ${totalventa} moneda corriente COP")
totalventa2 = tienda2["Precio"].sum()
print(f"la venta en la tienda #2 es: ${totalventa2} moneda corriente COP")
totalventa3 = tienda3["Precio"].sum()
print(f"la venta en la tienda #3 es: ${totalventa3} moneda corriente COP")
totalventa4 = tienda4["Precio"].sum()
print(f"la venta en la tienda #4 es: ${totalventa4} moneda corriente COP")

costoVentaTotal = totalventa + totalventa2 + totalventa3 + totalventa4
print(f"El total de venta en las tiendas es:${costoVentaTotal} moneda corriente COP")
print(f"GRAFICO VENTA POR TIENDA")

ventaTotal = [totalventa,totalventa2,totalventa3,totalventa4]
puntosDeVenta = ['Tienda #1','Tienda #2','Tienda #3','Tienda #4']
colorBarra = ['#7fb3d5', '#717d7e','#f4d03f','#1abc9c']
plt.figure(figsize=(5,4))
barras = plt.bar(puntosDeVenta, ventaTotal, color=colorBarra)
plt.title('VENTAS POR TIENDA')
plt.xlabel('Tiendas')
plt.ylabel('Valor en moneda corriente COP')
plt.show()


```

# 2. Ventas por categoría

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
puntosDeVenta = ['Tienda #1', 'Tienda #2', 'Tienda #3', 'Tienda #4']

for puntosDeVenta, tienda in zip(puntosDeVenta, tiendas):
    print(f"\n{puntosDeVenta}:\n")

    ventasporcategoria = tienda.groupby('Categoría del Producto').agg(
        Unidades_Vendidas=('Precio', 'count'),
        Venta_Total_COP=('Precio', 'sum')
   ).reset_index().sort_values(by='Unidades_Vendidas', ascending=False)
    print(ventasporcategoria)

    plt.figure(figsize=(5, 3))
    plt.pie(
        ventasporcategoria['Unidades_Vendidas'],
        labels=[f"{cat} ({cant})" for cat, cant in zip(ventasporcategoria['Categoría del Producto'], ventasporcategoria['Unidades_Vendidas'])],
        autopct='%1.1f%%',
        startangle=90,
        shadow=True
    )
    plt.title(f'VENTA DE PRODUCTO POR CATEGORIA - {puntosDeVenta}')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

print("Análisis: se puede establecer de acuerdo a las gráficas estadísticas,que, de acuerdo al nivel de venta por producto, no siempre el mas vendido es el que aporta el mayor ingreso por venta. Para este caso los electrodomésticos muestran un top mayor de ingreso económico con respecto al ítem de muebles.")
```

# 3. Calificación promedio de la tienda


```python
PromedioTienda1 = round(tienda["Calificación"].mean(), 2)
PromedioTienda2 = round(tienda2["Calificación"].mean(), 2)
PromedioTienda3 = round(tienda3["Calificación"].mean(), 2)
PromedioTienda4 = round(tienda4["Calificación"].mean(), 2)
print(f"Calificación promedio obtenida por la Tienda #1: {PromedioTienda1}")
print(f"Calificación promedio obtenida por la Tienda #2: {PromedioTienda2}")
print(f"Calificación promedio obtenida por la Tienda #3: {PromedioTienda3}")
print(f"Calificación promedio obtenida por la Tienda #4: {PromedioTienda4}")
print("El rango de promedio de calificación de las tiendas se encuentra por encima de los 3,9 puntos.")

promedio = [PromedioTienda1,PromedioTienda2,PromedioTienda3,PromedioTienda4]

promedioTotal = [PromedioTienda1,PromedioTienda2,PromedioTienda3,PromedioTienda4]
puntosDeVenta = ['Tienda #1','Tienda #2','Tienda #3','Tienda #4']
colorBarra = ['#7fb3d5', '#717d7e','#f4d03f','#1abc9c']
plt.figure(figsize=(5,4))
barras = plt.bar(puntosDeVenta, promedio, color=colorBarra)
plt.title('CALIFICACIÓN PROMEDIO OBTENIDA POR LA TIENDA')
plt.xlabel('Tiendas')
plt.ylabel('Promedio')

for i, promedio in enumerate(promedio):
    plt.text(i, promedio + promedio * 0.01, f"{int(promedio):,}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Agrupar por producto y sumar la cantidad vendida
ventas1 = tienda.groupby('Producto')['Precio'].sum()
ventas2 = tienda2.groupby('Producto')['Precio'].sum()
ventas3 = tienda3.groupby('Producto')['Precio'].sum()
ventas4 = tienda4.groupby('Producto')['Precio'].sum()

# Unificar en un DataFrame
df_ventas = pd.DataFrame({
    'Tienda 1': ventas1,
    'Tienda 2': ventas2,
    'Tienda 3': ventas3,
    'Tienda 4': ventas4
}).fillna(0)

# Preparar datos para graficar
productos = df_ventas.index.tolist()
x = np.arange(len(productos))
ancho = 0.2

# Gráfico de barras agrupadas
plt.figure(figsize=(35, 8))

for i, tienda in enumerate(df_ventas.columns):
    plt.bar(x + i * ancho, df_ventas[tienda], width=ancho, label=tienda)

plt.xticks(x + ancho * 1.5, productos, rotation=45)
plt.xlabel("Productos")
plt.ylabel("Cantidad Vendida")
plt.title("Comparación de Ventas por Producto entre Tiendas")
plt.legend()
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular promedio y cantidad de envíos
datos_envios = {
    'Tienda 1': {
        'promedio': tienda['Costo de envío'].mean(),
        'cantidad': tienda['Costo de envío'].count()
    },
    'Tienda 2': {
        'promedio': tienda2['Costo de envío'].mean(),
        'cantidad': tienda2['Costo de envío'].count()
    },
    'Tienda 3': {
        'promedio': tienda3['Costo de envío'].mean(),
        'cantidad': tienda3['Costo de envío'].count()
    },
    'Tienda 4': {
        'promedio': tienda4['Costo de envío'].mean(),
        'cantidad': tienda4['Costo de envío'].count()
    }
}

# Mostrar en consola
for tienda, datos in datos_envios.items():
    print(f"{tienda}:")
    print(f"  → Promedio de envío: ${datos['promedio']:.2f}")
    print(f"  → Cantidad de envíos: {datos['cantidad']}\n")

# Preparar datos para gráfico
tiendas = list(datos_envios.keys())
promedios = [datos_envios[t]['promedio'] for t in tiendas]
cantidades = [datos_envios[t]['cantidad'] for t in tiendas]

# Gráfico con doble eje Y
fig, ax1 = plt.subplots(figsize=(10, 6))

color1 = 'tab:blue'
ax1.set_xlabel('Tiendas')
ax1.set_ylabel('Promedio del envío ($)', color=color1)
ax1.bar(tiendas, promedios, color=color1, alpha=0.7, label='Promedio del envío')
ax1.tick_params(axis='y', labelcolor=color1)

# Segundo eje Y para cantidad de envíos
ax2 = ax1.twinx()
color2 = 'tab:orange'
ax2.set_ylabel('Cantidad de envíos', color=color2)
ax2.plot(tiendas, cantidades, color=color2, marker='o', linewidth=2, label='Cantidad de envíos')
ax2.tick_params(axis='y', labelcolor=color2)

plt.title("Promedio y Cantidad de Envíos por Tienda")
fig.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
